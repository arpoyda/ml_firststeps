## II. Indexing and Selecting:

```python
df = pd.read_csv('./source/winemag-data-130k-v2.csv', index_col=0)
```

- __Getting a column and row from DataFrame:__ <br>
_Remember: Columns and rows are just Series_

```python
# To get column use [name of column] or [list(names of columns)]
col = df['country']
cols = df[['country', 'region_1']]

# To get row use names: .loc[list(names of rows)]
row_1 = df_csv.loc['Portugal'] # if several rows named 'Portugal' you will get all of them
rows_1 = df_csv.loc[['Portugal', 'US']]

# or num_indexes: .iloc[list(num_indexes of rows)]
row_2 = df_csv.iloc[1]
rows_2 = df_csv.iloc[[1, 2, 3]]

# or slices: .iloc[n1:n2] or just [n1:n2]
rows_3 = df_csv.iloc[11:24]
rows_3 = df_csv[11:24]
```

__P.S.__ Series can be interpreted as a DataFrame with only one column (without name) and many rows (with names in original DataFrame). Name of the original column is the name of Series now.

__remember:__ almost every operations can be used in chain style:

```python
_ = df['country'][1:15]
```

<br>

- __Selecting data:__<br>
To find cells which have certain properties use boolean expresions in square breackets:<br>
or -> |, and -> &

```python
_ = df[(df['country'] == 'Italy') | (df['points'] == 86)]
_ = df[(df['country'] == 'Italy') & (df['points'] > 88)]
_ = df[df['country'].isin(['Italy', 'US', 'Portugal'])]
```

## III. Some functions and Map/Apply

- __There are several useful functions:__<br>
.mean(): return mean value of Series<br>
.median(): return median of Series<br>
.unique(): return all unique values of Series<br>
.idxmax(): return id of max element<br>

```python
_ = df['points'].mean()
_ = df['points'].median()
_ = df['points'].unique()
_ = df['points'].idxmax()
```

 - __Operators__ [+, -, *, /] __are being applied for whole Series:__

```python
_ = df['points'] + 1 
_ = df['points'] - 1
_ = df['points'] * 1
_ = df['points'] / 1
```

- __.map(func) applies__ _func_ __for every cell of Series and return new Series:__

```python
def func(cell):
    return cell*2 - 1

_ = df['points'].map(func)
```

Role of _func_ can be perfomed by _lambda_:

```python
_ = df['points'].map(lambda x: x*2 - 1)
```

 - __.apply(func, axis=) applies__ _func_ __for every row/column:__<br>
    axis = 0 or ‘index’: apply function to each column.<br>
    axis = 1 or ‘columns’: apply function to each row.


```python
def func(row):
    return str(row['points']) + str(row['country'])

_ = df.apply(func, axis='columns')
```
